## Rolled in one func dominance preprocessing

```python
import pandas as pd

dom = pd.read_csv('/content/BitcoinDominnance2018-24.csv') # format YYYY/MM/DD
dom.columns = ['Date','Value']
dom['Date'] = pd.to_datetime(dom['Date']).dt.strftime('%Y-%m-%d')
dom.shape
dom.loc['Date'] = dom['Date'].str.replace('/', '-')
dom = dom.dropna()

df = dom
expanded_df = pd.DataFrame()

for i in range(len(df)-1):
    start_date = df.iloc[i,0]
    end_date   = df.iloc[i+1,0]
    start_val  = df.iloc[i,1]
    end_val    = df.iloc[i+1,1]
    dates = pd.date_range(start=start_date, end=end_date, freq='D')
    values = np.linspace(start_val, end_val, len(dates))
    temp = pd.DataFrame({
        "DATE_STR": dates,
        "Value": values
    })
    expanded_df = pd.concat([expanded_df, temp])
expanded_df = expanded_df.drop_duplicates(subset="DATE_STR")

# Reset index
expanded_df = expanded_df.reset_index(drop=True)
expanded_df['DATE_STR'] = pd.to_datetime(expanded_df['DATE_STR']).dt.strftime('%Y-%m-%d')

# Save to new Excel
# expanded_df.to_excel("expanded_file.xlsx", index=False)

# print("Expanded Excel file saved as expanded_file.xlsx")
```

## Rolled in one function main df

```python
def make_csv(csv1_path, csv2_path):
  hr_by_hr = pd.read_csv(csv1_path)
  f_a_g = pd.read_csv(csv2_path)
  hr_by_hr['Fear_Greed_Index'] = np.nan
  hr_by_hr['Sentiment'] = np.nan
  # hr_by_hr['DATE_STR'] = pd.to_datetime(hr_by_hr['DATE_STR']).dt.date
  # f_a_g['Date'] = pd.to_datetime(f_a_g['Date']).dt.date
  # merge  = pd.merge(
  #     hr_by_hr,
  #     f_a_g[['Date' ,'Fear_Greed_Index', 'Sentiment']],
  #     left_on='DATE_STR',
  #     right_on='Date',
  #     how='left')
  # hr_by_hr = merged
  for i in f_a_g['Date']:
    hr_by_hr.loc[hr_by_hr['DATE_STR'] == i, 'Fear_Greed_Index'] = f_a_g[f_a_g['Date'] == i]['Fear_Greed_Index'].iloc[0]
    hr_by_hr.loc[hr_by_hr['DATE_STR'] == i, 'Sentiment'] = f_a_g[f_a_g['Date'] == i]['Sentiment'].iloc[0]

  #
  hr_by_hr['ma_24h'] = hr_by_hr['CLOSE_PRICE'].rolling(24).mean()
  hr_by_hr['ma_72h'] = hr_by_hr['CLOSE_PRICE'].rolling(72).mean()
  hr_by_hr['ma_168h'] = hr_by_hr['CLOSE_PRICE'].rolling(168).mean()

  # Percentage change with moving avg
  hr_by_hr['vol_8h'] = hr_by_hr['CLOSE_PRICE'].pct_change().rolling(8).std()# 8 hrs
  hr_by_hr['vol_24h'] = hr_by_hr['CLOSE_PRICE'].pct_change().rolling(24).std() # 1 day
  hr_by_hr['vol_72h'] = hr_by_hr['CLOSE_PRICE'].pct_change().rolling(72).std() # 3 days
  # Percentage change
  hr_by_hr['r_24h'] = hr_by_hr['CLOSE_PRICE'].pct_change(24) # 1 day
  hr_by_hr['r_72h'] = hr_by_hr['CLOSE_PRICE'].pct_change(72) # 3 day
  hr_by_hr['r_168h'] = hr_by_hr['CLOSE_PRICE'].pct_change(168) # 1 week
  # Close price change
  hr_by_hr['24_hr_close'] = hr_by_hr['CLOSE_PRICE'].pct_change(periods=24).shift(-24)
  # hr_by_hr['return_24h'] =  hr_by_hr['CLOSE_PRICE'].pct_change(periods=24).shift(-24)#  Remove this one

  hr_by_hr['TIME_UNIXs'] = pd.to_datetime(hr_by_hr['TIME_UNIX'], unit='s')
  hr_by_hr['dayofweek'] = hr_by_hr['TIME_UNIXs'].dt.dayofweek  # 0=Monday
  hr_by_hr['is_weekend'] = (hr_by_hr['TIME_UNIXs'].dt.dayofweek >= 5).astype(int)
  hr_by_hr['hour_sin'] = np.sin(2*np.pi*hr_by_hr['HOUR_STR']/24)
  hr_by_hr['hour_cos'] = np.cos(2*np.pi*hr_by_hr['HOUR_STR']/24)
  hr_by_hr['month'] = hr_by_hr['TIME_UNIXs'].dt.month


  hr_by_hr.dropna(inplace=True)
  return hr_by_hr
```

```python
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def evaluate(m, X, y, name="set"):
    pred = m.predict(X)
    rmse = np.sqrt(mean_squared_error(y, pred))
    mae = mean_absolute_error(y, pred)
    r2 = r2_score(y, pred)
    print(f"{name} -> RMSE: {rmse:.6f}  MAE: {mae:.6f}  R2: {r2:.4f}")
    return pred
```

```python
def r2_eval(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds), True  # (name, value, higher_is_better)
```

#LSTM STRAT

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
def create_sequences(dataset, time_step= 168):
  Xx, Yy = [], []
  for i in range(len(dataset) - time_step - 1):
      Xx.append(dataset.iloc[i:(i + time_step), 0].values)
      Yy.append(dataset[''].iloc[i + time_step, 0])
  return np.array(Xx), np.array(Yy)
```

```python
scaler_X = StandardScaler()
scaler_y = StandardScaler()
Xx =np.array(X)
Yy =np.array(Y)
Yy = Yy.reshape(-1, 1)
X_scaled = scaler_X.fit_transform(Xx)      # keep 2D shape
y_scaled = scaler_y.fit_transform(Yy).reshape(-1)  # shape -> (n_rows,)
```

```python
def create_sequences(X, y, time_steps=60):
    """
    X: 2D array (n_rows, n_features)
    y: 1D array (n_rows,)
    returns:
      X_seq -> (n_samples, time_steps, n_features)
      y_seq -> (n_samples,)
    """
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i : i + time_steps])
        ys.append(y[i + time_steps])   # predict the value immediately after the window
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_sequences(X_scaled, y_scaled, 168)
print("X_seq.shape:", X_seq.shape, "y_seq.shape:", y_seq.shape)
# Expected: X_seq.shape -> (len(df)-TIME_STEPS, TIME_STEPS, n_features)
```

```python
X_train, X_testval, y_train, y_testval = train_test_split(
    X_seq, y_seq, test_size=0.3, shuffle=True, random_state=42
)
```

```python
X_test, X_val, y_test, y_val = train_test_split(
    X_testval, y_testval, test_size=0.5, shuffle=True, random_state=42
)
```

```python
model= keras.models.Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(168, 23)))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
X_train.shape
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint_callback = ModelCheckpoint(
    filepath='best_lstm_model.keras',  # path to save the model
    monitor='val_loss',  # metric to monitor
    save_best_only=True, # save only the best model
    mode='min',          # minimize the monitored metric
    verbose=1            # print messages when saving
)
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=50, validation_data=(X_test, y_test), callbacks=[checkpoint_callback])
```
